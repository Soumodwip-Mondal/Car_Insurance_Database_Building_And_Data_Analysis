```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

Dataset Creation

```python
def create_dataset(set_path):
    n=1000000
    customer_id = np.arange(1, n+1)
    vehicle_id = np.arange(1, n+1)
    tenure = np.random.choice([1,2,3,4],size=n, p=[0.2,0.3,0.4,0.1])
    vehicle_value = np.full(n,100000)
    premium = tenure * 100
    dates = pd.date_range("2024-01-01","2024-12-31")
    purchase_dates = np.random.choice(dates, n)
    policy_start = purchase_dates + pd.Timedelta(days=365)
    policy_end = policy_start + pd.to_timedelta(tenure*365, unit="D")
    policy_df = pd.DataFrame({
    "Customer_ID": customer_id,
    "Vehicle_ID": vehicle_id,
    "Vehicle_Value": vehicle_value,
    "Premium": premium,
    "Policy_Purchase_Date": purchase_dates,
    "Policy_Start_Date": policy_start,
    "Policy_End_Date": policy_end,
    "Policy_Tenure": tenure
    })
    policy_df.to_csv(path_or_buf=set_path,index=False)
    return "All Set"
```

```python
create_dataset("policy_dataset.csv")
```

```python
policy_df['Policy_Start_Date']=pd.to_datetime(policy_df['Policy_Start_Date'])
policy_df['Policy_End_Date']=pd.to_datetime(policy_df['Policy_End_Date'])
```

```python
policy_df['Tenure_Days'] = (policy_df['Policy_End_Date'] - policy_df['Policy_Start_Date']).dt.days
```

```python
policy_df['Daily_Premium'] = policy_df['Premium'] / policy_df['Tenure_Days']
```

```python
cutoff = pd.to_datetime("2026-02-28")
```

```python
policy_df['Earned_Days'] = (
    policy_df.apply(lambda row: min(row['Policy_End_Date'], cutoff), axis=1) 
    - policy_df['Policy_Start_Date']
).dt.days.clip(lower=0)
```

```python
policy_df['Earned_Premium'] = policy_df['Earned_Days'] * policy_df['Daily_Premium']

earned_premium = policy_df['Earned_Premium'].sum()
earned_premium
```
